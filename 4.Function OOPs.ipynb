```python
def BMICal():
    num=int(input("Enter the BMI Intex:"))
    if(num<18.5):
        print("Under wieght")
        weight="Under wieght"
    elif(num<=24.9):
        print("Normal weight")
        weight="Normal weight"
    elif(num<29.9):
        print("Overweight")
        weight="Overweight"
    else:
        print("Very Overweight")
        weight="Very Overweight"
    return weight
```
