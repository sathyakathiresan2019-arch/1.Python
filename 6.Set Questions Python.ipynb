```python
def votelist():
    age=int(input("Enter the age:"))
    if(age<18):
      
        return "children"
    elif(age<35):
        
        return "Adult"
    elif(age<59):
      
        return "citizen"
    else:

        return "senior citizen"
```

```python
voters=votelist()
print(voters)
if voters=="children":
    print("Not eligible")
else:
    print("Eligible")
```
